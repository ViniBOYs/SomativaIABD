```python
# Importação de Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import numpy as np
```

```python
dados = pd.read_csv('dadosacoes.csv')
dados
```

```python
sns.boxplot(x=dados['nome ação'], y=dados['preço ação R$'])
plt.title('Nome da Acão x Preço da Ação')
plt.xlabel('Nome da Ação')
plt.ylabel('Preço da Ação')
plt.show()
```

```python
sns.boxplot(x=dados['nome ação'], y=dados['valor de mercado R$ -(Bilhões)'])
plt.title('Nome da Acão x Valor de Mercado')
plt.xlabel('Nome da Ação')
plt.ylabel('Valor de Mercado R$ - (Bilhões)')
plt.show()
```

```python
dados.info()
```

```python
dados.describe()
```

```python
# Lidando com Dados Nulos
dados['qtde cotas'].value_counts()
```

```python
# Substituição dos valores vazios por nan

dados = dados.replace('', np.nan)
```

```python
# Realizando a Soma dos Valores Vazios

print(dados.isnull().sum())
```

```python
data = pd.get_dummies(dados ,columns=['nome ação'], drop_first=True)
features = data[['valor de mercado R$ -(Bilhões)', 'preço ação R$', 'qtde cotas']]
kmeans = KMeans(n_clusters=4, random_state=0).fit(features)
data['Cluster'] = kmeans.labels_

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

for cluster in data['Cluster'].unique():
    cluster_data = data[data['Cluster'] == cluster]
    ax.scatter(cluster_data['preço ação R$'], cluster_data['valor de mercado R$ -(Bilhões)'], cluster_data['qtde cotas'], label=f'Cluster {cluster}')

plt.title('Agrupamento de Dados Utilizando o Algoritmo KMeans')
plt.xlabel('Preço da Ação')
plt.ylabel('Valor de Mercado (Bilhões)')
```
